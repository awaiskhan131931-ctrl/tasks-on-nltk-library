**Text normaliztion in nltk**
text normalization is a process in of convrting raw text into a clean,standard format, so it can be processed easily in by NLP models.

It combines multiple cleaning steps together into one pipeline.


**Steps of Text Normalization**

Raw Text > Lowercasing > Tokenization > Removing Punctuation > Removing Stop Words > Stemming / Lemmatization > Clean Normalized Text

**Complete Pipeline at Once**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError

def normalize_text(text):
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Tokenize
    tokens = word_tokenize(text)

    # Step 3: Remove Punctuation
    tokens = [word for word in tokens if word.isalpha()]

    # Step 4: Remove Stop Words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Step 5: Stemming
    ps = PorterStemmer()
    tokens = [ps.stem(token) for token in tokens] # Changed 'word' to 'token'

    return tokens

text = "Python is GREAT for Natural Language Processing!"
print(normalize_text(text))
# ['python', 'great', 'natur', 'languag', 'process']
```

**Stemming:**
Stemming is the process of reduceing a word to its/base from by removing suffixes and prefixes.

**Example**

running > run

**Why we use**

**Types of Stemmers in NLTK**

**Porter Stemming**
work in 5 fixed phases, each with set of suffix rule apply one ofter another
* **step 1** Plural removel

cats     → cat

caresses → caress

* **step 2** Past tense / -ing removal
walked   → walk

running  → run
* **step 3**     Y → I

happy    → happi

* **step 5**   Longer suffix removal

ational  → ate
fulness   → full
iveness   → ive

* **step 5** Final cleanup

probate  → probat

rate     → rate  (protected)

```python
import nltk
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words_to_stem = ["running", "played", "studies", "beautiful", "eating"]
for word in words_to_stem:
  print (f"{word}  >>>>>  {ps.stem(word)}")
```

**Lancaster Stemming**
Uses a lookup table of 120 rules and keeps looping until no more rules apply.

**Example walkthrough — "generously"**

Step 1: ends in "ly"   → remove → "generous"

Step 2: ends in "ous"  → remove → "gener"

Step 3: no more rules match → STOP

Result: "gener"   ← very aggressive!


```python
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()

words = ["running", "played", "studies", "beautiful"]

for word in words:
    print(word, "→", ls.stem(word))
```

**What is Part of speech tagging**
POS tagging (also called grammatical tagging) is the process of automatically assinging a grammatical labels to each word in the sentence based on its definition and context.


**Sample Example**

Input:  "The cat sat on the mat"

Output: [("The", DT), ("cat", NN), ("sat", VBD), ("on", IN), ("the", DT), ("mat", NN)]

**Noun, Verb & Adjective Identification**

Noun, Verb & Adjective Identification is the process of automatically finding and labelling word in a sentence that belong to this three major grammatical categories.

```python
import nltk
from nltk import word_tokenize,pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

sentence = "The brilliant scientist quickly discovered an amazing new planet"

tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

for word, tag in tags:
  print(f"{word:15} → {tag}")

# separate each category

nouns = [word for word, tag in tags if tag.startswith("NN")]
verbs = [word for word, tag in tags if tag.startswith("VB")]
adjectives = [word for word, tag in tags if tag.startswith("JJ")]

print("Nouns:", nouns)
print("Verbs:", verbs)
print("Adjectives:", adjectives)
print('\n\n\n')
```

**POS Tagging Techniques**
There is four major technique used for POS tagging, ranging from sample rules to advanced deef learning.

**Ruled based teqnique**

Uses manual writthen linguistic rules to assign POS tags on based on word pattern.

**How it works**

*IF word ends with "-ing"        → TAG as VBG (verb gerund)

IF word ends with "-ed"         → TAG as VBD (past tense verb)

IF word ends with "-ly"         → TAG as RB  (adverb)

IF word ends with "-tion"       → TAG as NN  (noun)

IF word starts with Capital     → TAG as NNP (proper noun)

IF word is "the/a/an"           → TAG as DT  (determiner)*


```python
def Ruled_based_teqnique(word):
  if word.endswith('ing'):
    return 'VBG'
  elif word.endswith('ed'):
    return 'VBD'
  elif word.endswith('ly'):
    return 'RB'
  elif word.endswith('tion'):
    return 'NN'
  elif word.istitle():
    return 'NNP'
  else:
    return 'NN'   # default

example_words =  ["running", "jumped", "quickly", "education", "London"]
for w in example_words:
    print(f"{w:15} → {Ruled_based_teqnique(w)}")
```

**name entity regocntion**
name entity regocnition in nltk is a natural language processing technique that automatically identifies and classifies "name entity " ----real word object like people ,place ,and organization, in text.

**NLTK achieves this through a 3-step pipeline:**

Raw Text → Tokenization → POS Tagging → NE Chunking → Labeled Entities.


**The 3-Step Pipeline Explained**

**Step 1 — Tokenization**
Splits text into individual words/tokens.

**Step 2 — POS Tagging**
Labels each token with its part of speech.

**Step 3 — NE Chunking**
Groups tokens and classifies them as named entities.





```python
text = """Elon Musk, the CEO of Tesla and SpaceX, announced a new project in Austin, Texas.
He met with Tim Cook from Apple at the conference held in New York on Monday."""

import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree # Import Tree for checking subtree types

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng') # Changed to specifically download the 'eng' version
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt_tab') # Added to resolve LookupError
nltk.download('maxent_ne_chunker_tab') # Added to resolve LookupError

# Tokenize text
tokenz = word_tokenize(text)
tags = pos_tag(tokenz)
tree   = ne_chunk(tags)
print('\n\n')
for subtree in tree:
    if isinstance(subtree, Tree):
        name  = " ".join(word for word, tag in subtree.leaves())
        label = subtree.label()
        print(f"{name} → {label}")
```

**Chunking**
chunking is the process of gruoping word into meaning ful phrase using POS tags.

**What is Chinking?**
Chinking is the opposite of chunking — you define what to remove/exclude from a chunk.



```python
import nltk
from nltk import word_tokenize, pos_tag, RegexpParser


text = "John is eating a delicious apple in the garden"
tokens = word_tokenize(text)
tags = pos_tag(tokens)

# CHUNKING  grab noun phrases
chunk_grammar = "NP: {<DT>?<JJ>*<NN.*>}"

# CHINKING — grab everything EXCEPT verbs
chink_grammar = """
NP:
    {<.*>+}
    }<VB.*>{
"""

print("=== CHUNKING ===")
chunk_parser = RegexpParser(chunk_grammar)
print(chunk_parser.parse(tags))

print("\n=== CHINKING ===")
chink_parser = RegexpParser(chink_grammar)
print(chink_parser.parse(tags))

```

sentiment analysis is a technique in natural language processing(NLP) that aotumatically identifies and extract the emotional tone or opinion expressed taxt.

The core idea to determine whether a piece of  positive, negative or natural and some time more nuanced emotion like anger , joy, surprised, or disgust.

 * **Finance:** gauging market sentiment from news article or earning calls.

**How it works (approaches):**

* **Rule-based:** uses predefined word lists (e.g., "excellent" = positive, "terrible" = negative) with scoring

* **Deep learning / LLMs:**modern transformer models (like BERT or GPT) understand context much better, catching sarcasm, nuance, and complex phrasing


**Eample**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    score = sia.polarity_scores(text)['compound']

    if score >= 0.05:
        return "Positive 😊", score
    elif score <= -0.05:
        return "Negative 😞", score
    else:
        return "Neutral 😐", score

# Test it
samples = [
    "NLTK is incredibly powerful and easy to use!!!",
    "I can't believe how bad this is...",
    "The tutorial was informative."
]

for s in samples:
    label, score = analyze_sentiment(s)
    print(f"{label} ({score:.3f}) — {s}")
```

**Information extraction:**
Information extraction **(IE)** is the process of automatically extraction structured information from unstructured or semi-structured text.
**NLTK** provide several tools to perform IE tasks.

**Information Extraction Pipeline**

Raw Text
    >>>
Sentence Segmentation
    >>>
Tokenization
    >>>
POS Tagging
    >>>
Named Entity Recognition (NER)
    >>>
Relation Extraction
    >>>
Structured Information

**Complete IE Pipeline Example**

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)
nltk.download('maxent_ne_chunker_tab', quiet=True) # Added to resolve LookupError for ne_chunker

def information_extraction_pipeline(text):
    print("=== STEP 1: Sentence Segmentation ===")
    sentences = sent_tokenize(text)
    print(sentences)

    print("\n=== STEP 2: Tokenization ===")
    tokens = word_tokenize(text)
    print(tokens)

    print("\n=== STEP 3: POS Tagging ===")
    pos_tags = nltk.pos_tag(tokens)
    print(pos_tags)

    print("\n=== STEP 4: Named Entity Recognition ===")
    chunks = nltk.ne_chunk(pos_tags)
    for chunk in chunks:
        if hasattr(chunk, 'label'):
            entity = ' '.join(c[0] for c in chunk)
            print(f"{chunk.label()}: {entity}")

# Run pipeline
text = "Steve Jobs co-founded Apple Inc. in Cupertino, California in 1976."
information_extraction_pipeline(text)
```

**Entity extraction:**
Entity extraction (also called Named Entity Recignition) is the process of identifying and classifying named entity in text such as people, organiztion, location and dates etc.

| Types of Named Entities | | |
|:---|:---|:---|
| **Entity Type** | **Label** | **Example** |
| Person | PERSON | Steve Jobs, Obama |
| Organization | ORGANIZATION | Apple, NASA, WHO |
| Location | GPE | New York, Pakistan |
| Date | DATE | January 2020 |
| Time | TIME | 3:00 PM |
| Money | MONEY | $500 million |
| Facility | FACILITY | Eiffel Tower |
| Percentage | PERCENT | 20% |


```python
from nltk.tokenize import sent_tokenize

def extract_all_entities(text):
    sentences = sent_tokenize(text)
    all_entities = []

    for sent in sentences:
        tokens = word_tokenize(sent)
        pos_tags = nltk.pos_tag(tokens)
        chunks = nltk.ne_chunk(pos_tags, binary=False)

        for chunk in chunks:
            if hasattr(chunk, 'label'):
                entity = ' '.join(c[0] for c in chunk)
                label = chunk.label()
                all_entities.append((entity, label))

    return all_entities

text = """
Barack Obama was born in Hawaii.
He served as President of the United States.
He studied at Harvard University in Cambridge.
"""

entities = extract_all_entities(text)
for entity, label in entities:
    print(f"{label:15} → {entity}")
```
