**Mian types of tokenization**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt') # Ensure 'punkt' is downloaded for other potential uses, though not strictly for this specific fix.

text = "Hello, how are you?"
tokens = word_tokenize(text, preserve_line=True)
# Expected output: ['Hello', ',', 'how', 'are', 'you', '?']
print(tokens)
```

**lowercassing in NLTK**
Lowercassing mean converting all text/tokens to lowercase so that words like "Apple", "APPLE", and "apple" are treated as the same word.


**why we use lowercasing**
1.   Avoid dupicated word with different cases.
2.   Reduce vocabulary size.
3.   Makes text consistent for processing.
2.   Improving accuracy of NLP models.

**Example in NLTK**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
text = "Python is GREAT for Natural Language Processing!"

tokens = word_tokenize(text, preserve_line=True)

lowercased_tokens = [token.lower() for token in tokens]

print(lowercased_tokens)
# Expected output: ['python', 'is', 'great', 'for', 'natural', 'language', 'processing']
```

**why we use Removing punctuation**


*   Puntuation has no meaning in most NLP tasks.

**Text normaliztion in nltk**
text normalization is a process in of convrting raw text into a clean,standard format, so it can be processed easily in by NLP models.

It combines multiple cleaning steps together into one pipeline.


**Steps of Text Normalization**

Raw Text > Lowercasing > Tokenization > Removing Punctuation > Removing Stop Words > Stemming / Lemmatization > Clean Normalized Text

**Complete Pipeline at Once**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

def normalize_text(text):
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Tokenize
    tokens = word_tokenize(text)

    # Step 3: Remove Punctuation
    tokens = [word for word in tokens if word.isalpha()]

    # Step 4: Remove Stop Words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Step 5: Stemming
    ps = PorterStemmer()
    tokens = [ps.stem(word) for word in tokens]

    return tokens

text = "Python is GREAT for Natural Language Processing!"
print(normalize_text(text))
# ['python', 'great', 'natur', 'languag', 'process']
```

**Stemming:**
Stemming is the process of reduceing a word to its/base from by removing suffixes and prefixes.

**Example**

running > run
playing > play
beautiful > beauti
studies > studi

**Why we use**


*   Reduce vocabulary size
* Treats similer word as same as same word.
* Makes NLP models faster.
* Help in the search engine

**Types of Stemmers in NLTK**

**Porter Stemming**
work in 5 fixed phases, each with set of suffix rule apply one ofter another
* **step 1** Plural removel

cats     → cat

caresses → caress

* **step 2** Past tense / -ing removal
walked   → walk

running  → run
* **step 3**     Y → I

happy    → happi

* **step 5**   Longer suffix removal

ational  → ate
fulness   → full
iveness   → ive

* **step 5** Final cleanup

probate  → probat

rate     → rate  (protected)

**Lancaster Stemming**
Uses a lookup table of 120 rules and keeps looping until no more rules apply.

**Example walkthrough — "generously"**

Step 1: ends in "ly"   → remove → "generous"

Step 2: ends in "ous"  → remove → "gener"

Step 3: no more rules match → STOP

Result: "gener"   ← very aggressive!


```python
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()

words = ["running", "played", "studies", "beautiful"]

for word in words:
    print(word, "→", ls.stem(word))
```

**3. Snowball Stemmer — Improved Region-Based Rules**

Work sammeler to the Porter but introduce Two region in the word to make rules smarter.

*R1 = region after the first non-vowel following a vowel*

*R2 = R1 applied again inside R1*

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")

words = ["running", "played", "studies", "beautiful"]

for word in words:
    print(word, "→", ss.stem(word))
```

**Lemmatization in NLTK**
Lemamtization in the process to reduce the word to its actual dictionary form (lemma) using vocabulary and grammer rules.

**WordNet Lemmatization **
WordNet Lemmatization is the process of reducing to its base or dictionary from (called lemma) using the WordNet lexical database.



```python
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

words = ["cats", "cacti", "geese", "rocks", "running", "better"]
for word in words:
    print(f"{word} → {lemmatizer.lemmatize(word)}")
```

**Difference Between Stemming and Lemmatization**

**Stemming**

* cutoff suffixes blindly without understading meaning.
* Output is always a real word.
* Word very fast.
* less accurate.
* Does not use grammer rules.
* Does not check the dictionary.
* Best for large dataset and search engine.
* **Example** "studies" → "studi"

**Lemmatization**

* Find the actual dictionary form of a word.
* Output is always a real word.
* Work slower then stemming.
* more accurate.
* Uses gramer rules and POS rules.
* Check the dictionary.
* Best for Chatbot and AI.
* **Example** "studies" → "study"
