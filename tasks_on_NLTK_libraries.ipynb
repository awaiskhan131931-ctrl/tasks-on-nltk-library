**What is NLTK?**
(natural language toolkit) this is python library working for human language data (text).
think of nltk is as a swiss army for text processing - its contain every think you need to start building natural processing (NLP) application.

**tools of NLTK**

**Tokenization**
tokenization is the breakdown of text into smaller unit called token.
these tokens can be word sentence ,punctuation marks,number and other meaningfull element.

in nltk tekenization is the fundamental first step in text preprocessing for NLP task.

**Mian types of tokenization**

1.**Word tokenization:** split text into indivdual word and punctuation.

2.**Sentence tokenization:** Split text into indevidual sentance


```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt') # Ensure 'punkt' is downloaded for other potential uses, though not strictly for this specific fix.

text = "Hello, how are you?"
tokens = word_tokenize(text, preserve_line=True)
# Expected output: ['Hello', ',', 'how', 'are', 'you', '?']
print(tokens)
```

```python
import nltk
from nltk.tokenize.punkt import PunktSentenceTokenizer
from nltk.data import find

nltk.download('punkt') # Ensure punkt is downloaded

text = "I love Python. It is great. NLTK is powerful."

# Explicitly load the 'punkt' model and use PunktSentenceTokenizer
punkt_path = find('tokenizers/punkt/english.pickle')
tokenizer = PunktSentenceTokenizer(punkt_path)
sentences = tokenizer.tokenize(text)

print(sentences)
# Expected output: ['I love Python.', 'It is great.', 'NLTK is powerful.']
```

**lowercassing in NLTK**
Lowercassing mean converting all text/tokens to lowercase so that words like "Apple", "APPLE", and "apple" are treated as the same word.


**why we use lowercasing**
1.   Avoid dupicated word with different cases.
2.   Reduce vocabulary size.
3.   Makes text consistent for processing.
2.   Improving accuracy of NLP models.

**Example in NLTK**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
text = "Python is GREAT for Natural Language Processing!"

tokens = word_tokenize(text, preserve_line=True)

lowercased_tokens = [token.lower() for token in tokens]

print(lowercased_tokens)
# Expected output: ['python', 'is', 'great', 'for', 'natural', 'language', 'processing']
```

**Removing punctuation**

Removing punctuation mean eliminating special cherecters like  from tokens so only meaningfun word remaining.

**why we use Removing punctuation**


*   Puntuation has no meaning in most NLP tasks.
*   Reduce noise in the data.
*   Makes text cleaner for processing.





```python
import nltk
import string
from nltk.tokenize import word_tokenize

nltk.download('punkt') # Ensure punkt is downloaded if not already

text = "Hello! How are you? I'm fine, thanks."

tokens = word_tokenize(text, preserve_line=True)

# Filter out punctuation tokens and convert to lowercase
clean_tokens = [word.lower() for word in tokens if word not in string.punctuation]

print(clean_tokens)
```

**Removing Stop Words in NLTK**
Stopword are common word that appear frequently in text but carry **little or no meaning** for NLP.

**Examples of Stop Words**

"is", "the", "a", "an", "in", "on", "at", "I", "we", "are", "was", "were"

**Why we remove stopword**


*   They add no value to the meaning .
*   Reduce nois to the data .
*   Makes NLP models faster and more accurate
* Reduce vocabulary size




```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
print(stop_words)
# {'i', 'me', 'my', 'myself', 'we', 'our', 'is', 'the', 'a', ...}

print("Total Stop Words:", len(stop_words))
# Total Stop Words: 179
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

text = "I am learning Natural Language Processing with NLTK"

# Tokenize
tokens = word_tokenize(text.lower(), preserve_line=True)
print("Before:", tokens)
# ['i', 'am', 'learning', 'natural', 'language', 'processing', 'with', 'nltk']

# Remove Stop Words
stop_words = set(stopwords.words('english'))
clean_tokens = [word for word in tokens if word not in stop_words]
print("After:", clean_tokens)
# ['learning', 'natural', 'language', 'processing', 'nltk']
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt') # Ensure punkt is downloaded
nltk.download('stopwords') # Ensure stopwords is downloaded

text = "I am learning Natural Language Processing with NLTK!"

tokens = word_tokenize(text.lower(), preserve_line=True)

stop_words = set(stopwords.words('english'))

# Remove both stop words and punctuation
clean_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

print(clean_tokens)
# ['learning', 'natural', 'language', 'processing', 'nltk']
```

**Text normaliztion in nltk**
text normalization is a process in of convrting raw text into a clean,standard format, so it can be processed easily in by NLP models.

It combines multiple cleaning steps together into one pipeline.


**Steps of Text Normalization**

Raw Text > Lowercasing > Tokenization > Removing Punctuation > Removing Stop Words > Stemming / Lemmatization > Clean Normalized Text

**Complete Pipeline at Once**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

def normalize_text(text):
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Tokenize
    tokens = word_tokenize(text)

    # Step 3: Remove Punctuation
    tokens = [word for word in tokens if word.isalpha()]

    # Step 4: Remove Stop Words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Step 5: Stemming
    ps = PorterStemmer()
    tokens = [ps.stem(word) for word in tokens]

    return tokens

text = "Python is GREAT for Natural Language Processing!"
print(normalize_text(text))
# ['python', 'great', 'natur', 'languag', 'process']
```
